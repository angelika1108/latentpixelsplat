```python
import torch

ckpt_path = "/home/angelika/latentpixelsplat/pretrained_models/acid.ckpt"
# ckpt_path = "/home/angelika/latentpixelsplat/pretrained_models/re10k.ckpt"

ckpt = torch.load(ckpt_path, map_location="cpu")
print(ckpt.keys())

for key in ckpt['state_dict'].keys():
    if ('backbone' not in key) and ('epipolar_transformer' not in key):
        print(key)
```

```python
ckpt_latent_path = '/home/angelika/vae/pretrained_models/model_vq-f4-noattn.ckpt'
# ckpt_latent_path = '/home/angelika/vae/pretrained_models/model_vq-f4.ckpt'

ckpt_latent_dec = torch.load(ckpt_latent_path, map_location="cpu")

print(ckpt_latent_dec.keys())

for key in ckpt_latent_dec['state_dict'].keys():
    if ('decoder' not in key) and ('encoder' not in key) and ('loss' not in key):
        print(key)
```

```python
key_list = []

for key in ckpt_latent_dec['state_dict'].keys():
    if ('encoder' in key) or ('loss' in key) or (key == 'quant_conv.weight') or (key == 'quant_conv.bias'):
        key_list.append(key)

for key in key_list:
    del ckpt_latent_dec['state_dict'][key]
```

```python
ckpt_latent_dec['state_dict'].update({'decoder.post_quant_conv.weight': ckpt_latent_dec['state_dict']['post_quant_conv.weight']})
ckpt_latent_dec['state_dict'].update({'decoder.post_quant_conv.bias': ckpt_latent_dec['state_dict']['post_quant_conv.bias']})
ckpt_latent_dec['state_dict'].update({'decoder.quantize.embedding.weight': ckpt_latent_dec['state_dict']['quantize.embedding.weight']})

del ckpt_latent_dec['state_dict']['post_quant_conv.weight']
del ckpt_latent_dec['state_dict']['post_quant_conv.bias']
del ckpt_latent_dec['state_dict']['quantize.embedding.weight']
```

```python
for key in ckpt_latent_dec['state_dict'].keys():
    ckpt['state_dict'].update({'decoder_latent.' + f'{key[8:]}': ckpt_latent_dec['state_dict'][key]})
    print(key)
```

```python
for key in ckpt['state_dict'].keys():
    if 'backbone' not in key and 'epipolar_transformer' not in key:
        print(key)
```
